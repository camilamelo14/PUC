```python
import collections

allkeys = r.keys('*')
a = []

for i in range(r.dbsize()):
    a.append(r.get(allkeys[i]))

result = collections.Counter(a)

print('Resultado:')
for word, count in result.most_common():
        print(word, count)
        
print('\nGanhador:',result.most_common()[0])
```
