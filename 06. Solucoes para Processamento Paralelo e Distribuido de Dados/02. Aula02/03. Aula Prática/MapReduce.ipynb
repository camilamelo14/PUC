### Map Reduce

- Patente original: Google
- A ideia deriva da programação funcional

#### Map: 
- Aplica a função/operação para cada elemento da lista (Uma multiplicação dos elementos por 2, por exemplo, não fax que os elesmentos tenham dependências dos resultados entre sí;
- Não altera o dado original >> Evita o princípio de 'Shared Data';
- Executa de forma paralela.

Detalhe do funcionamento:

- Um bloco no DFS possui vários 'membros' para processar os dados. CADA membro só pertence a apenas 1 bloco;
- Converte na estrutura cheva/valor
- Chaves não precisam ser únicas



#### Reduce: 
- Recebe os resultados das várias funções Map's
- Esta função recebe este resultado e gera uma agrupamento/erdução dos resultados, ex: Maior valor, valores somados etc.
- Se aplica sempre em uma lista
- Ex: [2,4,6,8] -> Reduce function (SOMA) -> [20]
- Previamente determinado
- O 'Name Node'controla o processo de gerenciamento dos Reduces
- Após todas tarefas map serem finalizadas, o 'master node' faz um merge dos arquivos.


##### Detalhes do funcionamento

- Armazena um arquivo no DFS
- Um conj. de tarefas Map é criado, para cada Map existe um ou mais blocos que serão processados
- O resultado do Map é um elemento chave e valor ou tuplas
- As estruturas chave e valor são coletadas pelo master e ordenadas pelas chaves
- Agrupa-se as chaves e divide-as pelas tarefas do tipo Reduce
- As tarefas so tipo Reduce vãão agrupar os dados pelas chaves

- 'Name Node' é responsável pela criação das tarefas de Map e Reduce nos 'datas nodes'
- Boa prática, criar uma tarefa de map para cada bloco
- Reduza os números de Reduce para evitar a explosão de arquivos gerados pela tarefa map
