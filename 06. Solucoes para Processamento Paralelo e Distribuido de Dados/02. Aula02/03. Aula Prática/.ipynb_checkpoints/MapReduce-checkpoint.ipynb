### Map Reduce

- Patente original: Google
- A ideia deriva da programao funcional

#### Map: 
- Aplica a funo/operao para cada elemento da lista (Uma multiplicao dos elementos por 2, por exemplo, no fax que os elesmentos tenham dependncias dos resultados entre s;
- No altera o dado original >> Evita o princpio de 'Shared Data';
- Executa de forma paralela.

Detalhe do funcionamento:

- Um bloco no DFS possui vrios 'membros' para processar os dados. CADA membro s pertence a apenas 1 bloco;
- Converte na estrutura cheva/valor
- Chaves no precisam ser nicas



#### Reduce: 
- Recebe os resultados das vrias funes Map's
- Esta funo recebe este resultado e gera uma agrupamento/erduo dos resultados, ex: Maior valor, valores somados etc.
- Se aplica sempre em uma lista
- Ex: [2,4,6,8] -> Reduce function (SOMA) -> [20]
- Previamente determinado
- O 'Name Node'controla o processo de gerenciamento dos Reduces
- Aps todas tarefas map serem finalizadas, o 'master node' faz um merge dos arquivos.


##### Detalhes do funcionamento

- Armazena um arquivo no DFS
- Um conj. de tarefas Map  criado, para cada Map existe um ou mais blocos que sero processados
- O resultado do Map  um elemento chave e valor ou tuplas
- As estruturas chave e valor so coletadas pelo master e ordenadas pelas chaves
- Agrupa-se as chaves e divide-as pelas tarefas do tipo Reduce
- As tarefas so tipo Reduce vo agrupar os dados pelas chaves

- 'Name Node'  responsvel pela criao das tarefas de Map e Reduce nos 'datas nodes'
- Boa prtica, criar uma tarefa de map para cada bloco
- Reduza os nmeros de Reduce para evitar a exploso de arquivos gerados pela tarefa map
