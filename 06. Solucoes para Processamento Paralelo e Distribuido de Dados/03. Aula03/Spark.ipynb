## Spark

-  plataforma (framework) para processamento paralelo e distribudo de propsito geral, utiliza bastante memria RAM.
- Extende os algoritmos padres de MapReduce. Ex: Join etc.
- APIs disponveis: Java, Scala, Python, SQL e R.

_Motivao_: 
Ambiente nico par axecuo de vrios tipos de processamentos:
- Batch
- Processamento interativos
- Processamento de streamings (O que o torno multi-propsito)



#### Funcioonalidades bsicas do Spark Core

- Agendamento das tarefas
- Ger. de memria
- Tolerncia a falhas
- Integrao com sistemas de armazenamento (O Spark no tem um sistema HDFS -> Nesta ideia o Hadoop entra como um aliado)
- APIs RDDs para lidar com os dados, RDDs (chave/valor)  uma coleo de itens distribudos no cluster e que podem ser manipulados de forma paralela.

##### Spark SQL

Componentes para lidar com dados estruturados, acesso via SQL (similar ao Hive do Hadoop) e suporta vrios tipos de dados Hive Tables, Parquet e arquivos JSON.

Pode mesclar SQL com outras operaes suportadas pelo RDDs (Pode usar Python e SQL por exemplo em uma mesma aplicao - chamamos isso de interoperabilidade)

##### Spark Streaming

Componente para lidar com Streams de dados, lida comm logs, ou fila de mensagens e se integra com o API Spark Core, fornece tambm escalabilidade, tolerncia a falhas, integraes com dados em memria e disco etc.

##### Spark MLib

Possui algoritimos tpicos de aprendizado de mquina, aplica alguns tipos: classificao, regresso, agrupamente, filtros colaborativos (recomendao). << Algoritimos clssicos de otimizao de ML

Mahout -> ML no Hadoop

Tambm realiza a importao de dados e avaliao de modelos.

##### Spark GraphX
Componente para manipulao de grafos (Abstrao matemtica).



#### Gerenciadores

- Spark  escalvel para milhares de ns em execuo (YARN e ApachMesos, para gerenciar os recursos)

## Histrico e Motivao

Iniciado em 2009, como foi feita pela equipe que trabalhou no Hadoop 1.0 eles j vinham com a bagagem dos problemas encontrados na ferramenta anterior.

    A ideia do MapReduce foi desenvolvida para que o Spark pudesse trabalhar com Grafos DAG direcionados aclicos (Ou seja, uma vez que h um ponto sempre vai existir um ponto final.
    


Indo na linha do MapReduce, ele  um DAG 



Muito conhecido por fazer leitura e escrita de formas intensa (como visto na img acima) temos mts I/O 

- Soluo Adotada no Spark

As duas etapas formam um DAG que  a generalizao do MapReduce, assim o DAG no conflita com o paradigma MapReduce. Originalmente o MapReduce tem dificuldades de paralelizar vrios jobs MapReduce.

A abordagem DAG resolve o problema de paralelizar vrios jobs MapReduce



No DAG tambm  possvel reutilizar um processamento j realizado anteriormente que esteja em memria, o que no Hadoop no d.

#### O que o Spark adota como soluo

- Aliado com a utilizao do dados reaproveitveis e em memria;
- Implementaes complementares tambm fazem uso da mesma ideia (YARN Hadoop 2.X)
- Exemplo: TEZ utilizado na HortonWorks.

- Todo processamento no Spark  executado como DAG;
- Usa dados em memria e economiza leitura em disco;
- Utiliza RDDs para compartilhar dados e aproveitar para no fazer escrita no disco ao final de cada etapa.



Atualmente foi transferido para a fundao Apach (2013)
