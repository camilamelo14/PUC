## Spark

- É plataforma (framework) para processamento paralelo e distribuído de propósito geral, utiliza bastante memória RAM.
- Extende os algoritmos padrões de MapReduce. Ex: Join etc.
- API´s disponíveis: Java, Scala, Python, SQL e R.

_Motivação_: 
Ambiente único par axecução de vários tipos de processamentos:
- Batch
- Processamento interativos
- Processamento de streamings (O que o torno multi-propósito)



#### Funcioonalidades básicas do Spark Core

- Agendamento das tarefas
- Ger. de memória
- Tolerância a falhas
- Integração com sistemas de armazenamento (O Spark não tem um sistema HDFS -> Nesta ideia o Hadoop entra como um aliado)
- API´s RDD´s para lidar com os dados, RDD´s (chave/valor) é uma coleção de itens distribuídos no cluster e que podem ser manipulados de forma paralela.

##### Spark SQL

Componentes para lidar com dados estruturados, acesso via SQL (similar ao Hive do Hadoop) e suporta vários tipos de dados Hive Tables, Parquet e arquivos JSON.

Pode mesclar SQL com outras operações suportadas pelo RDD´s (Pode usar Python e SQL por exemplo em uma mesma aplicação - chamamos isso de interoperabilidade)

##### Spark Streaming

Componente para lidar com Streams de dados, lida comm logs, ou fila de mensagens e se integra com o API Spark Core, fornece também escalabilidade, tolerância a falhas, integrações com dados em memória e disco etc.

##### Spark MLib

Possui algoritimos típicos de aprendizado de máquina, aplica alguns tipos: classificação, regressão, agrupamente, filtros colaborativos (recomendação). << Algoritimos clássicos de otimização de ML

Mahout -> ML no Hadoop

Também realiza a importação de dados e avaliação de modelos.

##### Spark GraphX
Componente para manipulação de grafos (Abstração matemática).



#### Gerenciadores

- Spark é escalável para milhares de nós em execução (YARN e ApachMesos, para gerenciar os recursos)

## Histórico e Motivação

Iniciado em 2009, como foi feita pela equipe que trabalhou no Hadoop 1.0 eles já vinham com a bagagem dos problemas encontrados na ferramenta anterior.

    A ideia do MapReduce foi desenvolvida para que o Spark pudesse trabalhar com Grafos DAG direcionados acílicos (Ou seja, uma vez que há um ponto sempre vai existir um ponto final.
    


Indo na linha do MapReduce, ele é um DAG 



Muito conhecido por fazer leitura e escrita de formas intensa (como visto na img acima) temos mts I/O 

- Solução Adotada no Spark

As duas etapas formam um DAG que é a generalização do MapReduce, assim o DAG não conflita com o paradigma MapReduce. Originalmente o MapReduce tem dificuldades de paralelizar vários jobs MapReduce.

A abordagem DAG resolve o problema de paralelizar vários jobs MapReduce



No DAG também é possível reutilizar um processamento já realizado anteriormente que esteja em memória, o que no Hadoop não dá.

#### O que o Spark adota como solução

- Aliado com a utilização do dados reaproveitáveis e em memória;
- Implementações complementares também fazem uso da mesma ideia (YARN Hadoop 2.X)
- Exemplo: TEZ utilizado na HortonWorks.

- Todo processamento no Spark é executado como DAG;
- Usa dados em memória e economiza leitura em disco;
- Utiliza RDDs para compartilhar dados e aproveitar para não fazer escrita no disco ao final de cada etapa.

# Arquitetura e conceito

###### Drive Program (Gestor de execução da aplicação)
Qualquer aplicação Spark é um "Drive Program", esta aplicação é o que faz com que as operações possam ser realizadas em paralelo no Cluster Spark e possui uma função principal.

Gerencia a alocação dos recursos para execução de uma aplicação.

###### Contexto Spark (Objeto utilizado pelo DP para conversar com o Cluster)
Representa a conexão da aplicação com o Cluster, usando o Spark Shell automaticamente um contexto é criado.
