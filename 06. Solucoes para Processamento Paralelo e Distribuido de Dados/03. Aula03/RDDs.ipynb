## SPARK RDD´s

- Coleção distriuída de elementos (Podemos considerar um grande objeto <chave> e <valor>
- Lógica de trabalho: Cria-se um RDD (Objeto no exemplo como lines.) -> Transforma o RDD -> Gerar resultado
- De forma transparente para nós o Spark distribui os dados do RDDs no cluster e paraleliza as operações.
    


RDD executado como fosse um DataFlow

RDD´s representam um conjunto distribuído e imutável, sendo involátil facilita o processo de tolerência a falhas e os RDDs são divididos em partições e estas partições podem ser estabalecidas por nós).

O RDD sendo imutável não é possível realizar um update em um RDD e esses dados permanecem em disco ou memória de forma indistinta.

O Driven Program é que coordena as atividades e os workers armazenam e manipulam as partições RDD´s



Ao criar um RDD duas operações são possíveis:

##### Transformations:
    
1. Executa dentro de um Worker
2. Criar um RDD apartir de um anterior
3. O resultado sempre gera um outro RDD

##### Actions:

1. Calcula algum valor sobre um RDD
2. O Resultado pode ser criado em um HDFS
3. Retorna uma medida *Não um RDD*
4. Por padrão todas as vezes que uma action é executada, todos os RDD´s relacionados a ela são recalculados. (Ou seja, toda action reezecuta as transformations existentes) -> A Importância de gravar dados em memória

### Conceitos

- Lazy Evaluation
    - As transformation são postergadas até encontrar uma action -> Ele faz isso para conhecer melhor o grafo da execução e adaptar melhor processamento e uso.
    - Todo programa Spark é um DAG
    


- Noções de persistência
    - Para executar uma RDD´s em várias actions um comando 'persist' deve ser explicitado
    - Uma vez que persistido os dados de uma RDD, ao executar uma actions o mesmo não é recalculado

```python
lines = sc.textFile("path/xpto")
lines.persist()

lines.persist().is_cached #Valida se está em memória

lines.count();
lines.frist()

lines.unpersist() #Não persistir os dados
```
