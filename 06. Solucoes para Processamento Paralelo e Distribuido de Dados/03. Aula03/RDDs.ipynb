## SPARK RDDs

- Coleo distriuda de elementos (Podemos considerar um grande objeto <chave> e <valor>
- Lgica de trabalho: Cria-se um RDD (Objeto no exemplo como lines.) -> Transforma o RDD -> Gerar resultado
- De forma transparente para ns o Spark distribui os dados do RDDs no cluster e paraleliza as operaes.
    


RDD executado como fosse um DataFlow

RDDs representam um conjunto distribudo e imutvel, sendo involtil facilita o processo de tolerncia a falhas e os RDDs so divididos em parties e estas parties podem ser estabalecidas por ns).

O RDD sendo imutvel no  possvel realizar um update em um RDD e esses dados permanecem em disco ou memria de forma indistinta.

O Driven Program  que coordena as atividades e os workers armazenam e manipulam as parties RDDs



Ao criar um RDD duas operaes so possveis:

##### Transformations:
    
1. Executa dentro de um Worker
2. Criar um RDD apartir de um anterior
3. O resultado sempre gera um outro RDD

##### Actions:

1. Calcula algum valor sobre um RDD
2. O Resultado pode ser criado em um HDFS
3. Retorna uma medida *No um RDD*
4. Por padro todas as vezes que uma action  executada, todos os RDDs relacionados a ela so recalculados. (Ou seja, toda action reezecuta as transformations existentes) -> A Importncia de gravar dados em memria

### Conceitos

- Lazy Evaluation
    - As transformation so postergadas at encontrar uma action -> Ele faz isso para conhecer melhor o grafo da execuo e adaptar melhor processamento e uso.
    - Todo programa Spark  um DAG
    


- Noes de persistncia
    - Para executar uma RDDs em vrias actions um comando 'persist' deve ser explicitado
    - Uma vez que persistido os dados de uma RDD, ao executar uma actions o mesmo no  recalculado

```python
lines = sc.textFile("path/xpto")
lines.persist()

lines.persist().is_cached #Valida se est em memria

lines.count();
lines.frist()

lines.unpersist() #No persistir os dados
```
