### Hadoop 1.0 vs Hadoop 2.0

- Hadoop 1.0
O NameNode: Guarda em memória RAM os metadados, Executa todas as operações relativas aos metadados, existe um ponto de falha da arquitetura pois apenas executa com um NameNode -> Máquina controladora (Citado em aula que esta falha é contornável com o uso do NameNode secundário)

Limitações:
- Escabilidade limitada de 4.000 máquinas/nós: gerando até 40.000 tarefas concorrentes.
- Gestão de recurso estática
- Não há suporte para os demais paradigmas de processamento -> Trabalhando apenas com o paradigma MapReduce



O professor sinalizou que como essas informações eram definidas previamente por este motivo não fazia o gerenciamento dinâmico desses recursos.

Muitas vezes existia um número alto de tarefas Maps, enquando poucas tarefas Reduce e essas ocupavam recursos no ambiente.



## Hadoop 2.0



Apenas uma máquina para gerenciar os recursos e outra máquina para acompanhar o ciclo de vida.

Neste conceito que aparece o YARN (YET ANOTHER RESOURCE NEGOTIATOR)



- Resource Manager: Responsável pela gestão de recursos
- Cada DataNode: Possui um JobTracker reduzido que fará a gestão do ciclo de vida do Job
- NodeManager: Gerentes do DataNode responsável por executar qq tipo de tarefa (Além do Map/Reduce)

Seguindo esta linha, o JOBTracker desaparece dando lugar para os ResourceManager



O NodeManager continua responsável por fazer a gestão das aplicações, é mais eficiente que o TaskTracker, faz qq tipo de computação no AM não somente o MapReduce

- Possui a ideia de Containers

    - Lida com vários resursos de RAM, CPU e IO e não exige um número fixo de funções Map´s e Reduce's
    - 0 Número de containers é limitado (Não pode execer os resursos da máquina)
