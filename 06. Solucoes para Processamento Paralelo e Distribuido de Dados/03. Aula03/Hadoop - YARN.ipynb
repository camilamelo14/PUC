### Hadoop 1.0 vs Hadoop 2.0

- Hadoop 1.0
O NameNode: Guarda em memria RAM os metadados, Executa todas as operaes relativas aos metadados, existe um ponto de falha da arquitetura pois apenas executa com um NameNode -> Mquina controladora (Citado em aula que esta falha  contornvel com o uso do NameNode secundrio)

Limitaes:
- Escabilidade limitada de 4.000 mquinas/ns: gerando at 40.000 tarefas concorrentes.
- Gesto de recurso esttica
- No h suporte para os demais paradigmas de processamento -> Trabalhando apenas com o paradigma MapReduce



O professor sinalizou que como essas informaes eram definidas previamente por este motivo no fazia o gerenciamento dinmico desses recursos.

Muitas vezes existia um nmero alto de tarefas Maps, enquando poucas tarefas Reduce e essas ocupavam recursos no ambiente.



## Hadoop 2.0



Apenas uma mquina para gerenciar os recursos e outra mquina para acompanhar o ciclo de vida.

Neste conceito que aparece o YARN (YET ANOTHER RESOURCE NEGOTIATOR)



- Resource Manager: Responsvel pela gesto de recursos
- Cada DataNode: Possui um JobTracker reduzido que far a gesto do ciclo de vida do Job
- NodeManager: Gerentes do DataNode responsvel por executar qq tipo de tarefa (Alm do Map/Reduce)

Seguindo esta linha, o JOBTracker desaparece dando lugar para os ResourceManager



O NodeManager continua responsvel por fazer a gesto das aplicaes,  mais eficiente que o TaskTracker, faz qq tipo de computao no AM no somente o MapReduce

- Possui a ideia de Containers

    - Lida com vrios resursos de RAM, CPU e IO e no exige um nmero fixo de funes Maps e Reduce's
    - 0 Nmero de containers  limitado (No pode execer os resursos da mquina)
