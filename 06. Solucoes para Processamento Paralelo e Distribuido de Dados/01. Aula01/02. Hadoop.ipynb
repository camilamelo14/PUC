###### HDFS (Hadoop distribute File System):
Sistema de arquivos distribudos para dados estruturados, semi-estruturados e no estruturados. Arquivos so divididos em bloco e armazenados com redundncia no cluster;

Caractersticas DFS(sistema de arquivo distribudo) so:

- Arquivos devem ser 'grandes', gigabytes, pelo menos; Arquivos menores no fazem sentido no DFS;
- Arquivos no DFS so raramente atualizados (write-once read-many). Adicionalmente dados so adicionados para os arquivos (periodicidade, processamento batch);
- Arquivos so divididos em partes ('chunks' ou blocos), normalmente 64 megabytes e replicados para, pelo menos, 3 ns (em racks diferentes);
- As informaes dos blocos e replicas  controlado utilizando metadados e com um figura central no cluster: 'name node' ou 'master node'; O Name node: Gerencia o sistema de arquivos(rplicas, blocos, ns e racks):abrir, fechar, renomear arquivos; Tambm gerencia o acesso dos clientes ao arquivos;
- Os outros ns do cluster so chamados de 'data node'ou 'slave node'; Executam as operaes enviados pelo 'Name node': criao, excluso e replicao de blocos;

###### Map Reduce:
Framework para execuo de processamento paralelos em mltiplos ns de trabalho para posteriormente combinar os resultados;

A implementao do algoritmo  utilizada para realizar computao no DFS para arquivos 'grandes' e comexecuo tolerante a falha.

O sistema lida com osdemais detalhes: 
- Execuo paralela; 
- Coordenao de tarefas (Map e reduce); 
- Lidar com a tolerncia a falhas;

###### YARN (Yet another resource negotiator) (Hadoop 2.0):
Gerenciamento da execuo das aplicaes no cluster;

O Yarn  gestor de recursos que foi criado pela comunidade e  usado no Hadoop 2.0 para melhorar o processo de gesto de recursos do Hadoop. Possui os seguintes componentes: 
- Resource Manage, 
- Application Manager e 
- Node Manager.

TaskTracker (sua verso anterior  no hadoop 1.0). Executa qualquer tipo de computao que faz sentido no contexto do Application Manager: no somente map e reduce. Possui o conceito de containers que so capazes de lidar com recursos variveis (RAM, CPU, IO) e no exigem nmero de funes maps e reduce pr definidas;

#### Demais componentes (exemplos) DATA SERVICES:

- Flume: Servio distribudo para coleta, agregao e movimentao de *streams* de logs de dados para o HDFS;
- Sqoop: (bidirecional) Movimentao de dados para dentro do Hadoop a partir de bancos de dados relacionais e vice versa;
- PIG: Possibilitar a escrita de programas de transformao, transformao e movimentao de dados utilizando uma linguagem simples de script (Pig Latin - linguagem) - Ferramenta de ETL;
- Hive: Interface SQL para o Hadoop que possibilita sumarizao de dados, consultas ad-hoc e anlise de grandes volumes de dados; (Arquivos CSV, Excell etc.) Linguagem HSQL
- HCatalogue; trabalha hoje junto com o Hive, e  o responsvel por criar os metadados
- HBASE: Banco de dados NoSQL(colunar -'big table clone') para utilizao de dados de forma interativa (no batch);
