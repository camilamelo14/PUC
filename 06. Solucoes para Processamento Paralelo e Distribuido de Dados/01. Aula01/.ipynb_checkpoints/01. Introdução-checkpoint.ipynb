#### Motivação

Devido o crescimento exponencial dos dados gerados atualmente há uma enorme necessidade de processamentos mais ágeis, neste vies são extremamente necessários computação mais rápida e grandes processamentos de bancos de dados.

A computação paralela e distribuída vem a tona para resolver estes problemas, sendo capaz de lidar com a quantidade de dados massiva e comporte escalabildiade elastástica.

#### 1.2 Arquitetura de soluções distribuídas

Algumas características interessantes de soluções paralelas:
- Conectividade ⇒ rede de interconexão
- Heterogeneidade ⇒ hardware e software distintos (possibilidade de ter S.O e Hardwares distintos em uma mesma rede)
- Compartilhamento ⇒ utilização de recursos (memória, disco, CPU)
- Imagem do sistema ⇒ como usuário o percebe
- Escalabilidade ⇒ mais estações de trabalho levam a melhor desempenho/eficiência

### Taxonomia de Flynn (Conceitual)

Uma das metodologias mais conhecidas e utilizadas para classificar uma arquitetura de computadores ou
conjunto de computadores é a __taxonomia de Flynn (1966)__. Duas dimensões: instruções e dados; Cada dimensão
assume dois valores distintos: single ou multiple



###### SISD – Single Instruction Single Data (Paradigma dos PCs atualmente)

- Arquitetura dos computadores com um __único processador__;
- Apenas uma instrução é processada a cada momento.
- Apenas um fluxo de dados é processado a cada momento.
- Exemplos: PCs, workstations e servidores com um único processador.



##### SIMD – Single Instruction Multiple Data (Necessário dados que contenham certo padrão)
- Tipo de arquitetura paralela desenhada para problemas específicos (alto padrão de regularidade nos dados, ex.:processamento de imagem);
- Todas as unidades de processamento executam a mesma instrução a cada momento;
- Cada unidade de processamento pode operar sobre um fluxo de dados diferente;
- __Exemplos__: Computadores com unidades de processadores gráficos Grafics Processors Units (GPUs);



##### MISD – Multiple Instruction Single Data (Dados com regularidade funcional)
- Arquitetura paralela desenhada para problemas caracterizados por um alto padrão de regularidade
funcional (ex.: processamento de sinal);
- Constituída por uma pipeline de unidades de processamento independentes que operam
sobre um mesmo fluxo de dados enviando os resultados de uma unidade para a próxima;
- Cada unidade de processamento executa instruções diferentes a cada momento;
- __Exemplos__:Não existem exemplos práticos



##### MIMD – Multiple Instruction Multiple Data (Capacidade explorada na matéria de processamento de dados)
- Arquitetura paralela predominante;
- Cada unidade de processamento executa instruções diferentes a cada momento;
- Cada unidade de processamento pode operar sobre um fluxo de dados diferente;
- __Exemplos__: alguns supercomputadores, clusters de computadores paralelos em rede, PCs
multi-core; 



#### Principais objetivos

Conceitualmente queremos melhorar o Speddup, segundo a lei de Amdahl $\frac{T(1)}{T(P)}$

Speedup esperado com número de processadores:

$$
\frac{1}{\frac{P}{N}+s}
$$

Sendo:

- P = Fração paralelizável
- N = Número de processadores
- S = Fração sequencial

__Exemplo__

- P = 0.7 S = 0.3 N=1 Speedup = 1
- P = 0.7 S = 0.3 N=2 Speedup = 1.53
- P = 0.7 S = 0.3 N=3 Speedup = 1.87



Plataformas MIMD podem ser agrupadas em quatro grupos: SMPs (Symmetric MultiProcessors), MPPs (Massively Parallel Processors), Cluster ou NOWs (Network Of Worstations) e Grades Computacionais

Será abordado em nossas aulas os produtos Hadoop e Spark, professor também comenta que há outros como Storm e Flink

### Plataformas MIMD - Multiple Instruction Multiple Data

##### SMPs ou Multiprocessadores



- Espaço de endereçamento (memória centralizada)
- Mecanismos de comunicação (Comunicação são feitas sincronizadas através de um espaço de endereçamento compartilhado (memória)
- Características:
    - Unício SO
    - Compartilhamento de memória
    - Fortemente acoplado
    - Não escalável (necessário adquirir recursos para melhorará-los)

##### MIMD - Multiple Instruction Multiple Data

- Espaço de endereçamento (memória centralizada)
- Mecanismos de comunicação (Comunicação são feitas sincronizadas através de um espaço de endereçamento compartilhado (memória)
- Agrupado em quatro grupos:
    - SMPs (Descrita acima)
    - MPPs
    - Cluster ou NOWs (Network Of Worstations)
    - Grades Computacionais

##### MPPs

- Espaço de endereçamento (Não compartilhado => Memória distribuída)
- Mecanismos de comunicação (__Troca de mensagens__, muito intensiva) Ex. TeraData
- Unidades de processamento (Multiplo processadores c/ memória privativa)
- Características:
    - Sistema homogêneo e heterogêneo
    - Cada nó executa seu S.O
    - Imagem única (A ótica do usuário final é que se está trabalhando em uma única máquina)
    - Rede de interconexões (dif. topologias)
    - Escaláveis
    - Apps não compartilham recursos em paralelo, podendo gerar tempo de espera
    
OBS: O Escalonador que defini o processamento que será dedicado a cada máquina 
    


##### Clusters  (MIMD)

- Espaço de endereçamento (Não compartilhado)
- Mecanismo comunicação (Troca de mensagens)
- Unidades de processamento (Conjuntos de estações de trabalho ou PCs)
- Características:
    - Sistema homogêneo e heterogêneo
    - Rede de interconexões (Redes locais, tendem a ser mais lentas que no MPP)
    - Cada nós com seu escalonador (Escaláveis)
    - Compoe um sistema de alto desempenho com baixo custo (quando comprados com MPP)
    


##### Grids

- Computadores independentes (qualquer um dos modelos apresentados podem compor este Grid computacional)
- Geograficamente distantes
- Não têm imagem única do sistema a princípio
- Pode ser formado por __PCs, SMPs, MPPs e clusters__.



### Conceitos Gerais

- Código Sequencial: Avaliar se o código são inerentemente sequenciais
- Concorrência: Número de tarefas concorrentes (desafio definição deste número de tarefa) 
- Granularidade: Qual tamanho da tarefa a ser executada? Ideal avaliar para evitar um overhead no SO.Ex: Uma máquina centralizada que executa este processamento, caso existam __milhões__ de pequenas tarefas provavelmente esta máquina apresentará problema ao executá-las
- Balanceamento de carga: Configurar adequadamente para que seja distribuído de forma uniforme

##### Memória Partilhada ou Distribuída

- Partilhada: Processors ou Threads (Memória partilhada)
- Distribuída: Usa troca de mensagens.

#### Principais paradigmas de programação paralela

##### Master/Slave (Forma mais trivial de programação paralela)

Divide a computação em duas entidades distintas:

- Master: é o responsável por decompor o problema em tarefas, distribuir as tarefas pelos slaves e recolher os resultados parciais dos slaves de modo a calcular o resultado final
- Slaves: responsabilidade trivais e simples: obter uma tarefa do master, processar a tarefa e enviar o resultado de volta para o master.



- Exemplo: Haddop é um exemplo deste paradigma, em relação ao balanceamento de carga o Haddop 1.0 era estático e no 2.0 o balanceamento pode ser dinâmico (IAR?)

Quando o número de Slave é muito alto para um Master pode gerar problemas (__Exemplos?->Proefessor 09:13 video Conceitos - Paradigmas de processamento paralelo)__.

- Haddop 1.0 e 2.0 --> Apenas um Master

###### Single Program Multiple Data (SPMD)

- Consiste em processos que executam o mesmo programa (executável) mas sobre diferentes partes dos dados.



- Depende de uma boa distribuíção das máquinas
- Existe uma comunicação entre as máquinas
- Maior escabildiade

__Muito sensível a falhas__ devido a comunicação entre as máquinas

###### Data Pipelining

- Utiliza uma decomposição funcional do problema em que cada processo executa apenas uma parte do algoritmo completo e total;
- O padrão de comunicação é definido e simples: processos são organizados em sequência (pipeline) e cada processo só troca informação com o processo seguinte;
- Dependente muito da capacidade de balancear a carga entre as diferentes etapas da pipeline;



Decompor o problema em várias soluções (filtro, executar algoritmos e etc.)

###### Divide and Conquer (Dividir para Conquistar)

Utiliza uma divisão recursiva do problema inicial em subproblemas independentes instâncias mais pequenas do problema inicial) cujos resultados são depois combinados para obter o resultado final



- Subtarefas totalmente independetes
- Comunicação existe apenas entre o processo que Pai e as subtarefas (filhos)

##### Speculative Parallelism

Não muito utilizado, faz de forma especulativa. (O que for retornado mais rápido será executado)
