# Spark

 uma plataforma (em cluster) para processamento __rpido, paralelo e de propsito geral que extende o algoritmo padro__: MapReduce. Muitos clculos so feitos em memria (podem ser executados no disco tambm).

 complementar ao Haddop

Possui APIs para:
- Java
- Scala
- Python
- SQL e
- R.



Conceitos importantes em uma aplicao Spark:

- __Driver Program__: Qualquer aplicao Spark  um 'Driver program'. Esta aplicao  capaz de fazer ope raes paralelas no cluster Spark.  como se fosse uma funo principal (main) e define/criadados distribudos no cluster e aplica operaes paralelas aos dados distribudos no cluster. As operaes no so apenas Map e Reduce;
- __Contexto Spark__: O 'Driver program' acessa o spark atravs de um objeto contexto que na verdade, representa a conexo da aplicao com o Cluster. Possui alguns parmetros de configurao.
- __Executors__: Para realizar as operaes do 'Driver Program', o mesmo faz uso de vrios ns que possuem os executores



#### RDDs

RDDs representam um conjunto distribudos de objetos 'imutveis' (no volatis).Isto facilita o processo de distribuio,  replicao e compartilhamento dos dados. Tais dados imutveis podem permanecer em disco ou memria de maneira indistinta.

A lgica de trabalho no Spark  criar RDDs, transformar RDDs existentes e executar operaes nos RDDs para gerar resultados. De forma transparente o Spark distribui os dados dos RDDs no cluster e paraleliza as operaes executadas nos RDDs

- RDDs so divididos em vrias 'partitions' que podem residir em vrios ns do Cluster

A criao de RDDs  feito de duas maneiras: 

- Carregando dados de forma externa, por exemplo, 
- ou distribuindo uma coleo de objetos no programa __Driver__



__O Driver program  quem coordena as atividades e os Workers armazenam e manipulam as parties dos RDDs__

Uma vez criado os RDDs, duas operaes so possveis:
- As __transformations__ criam um RDD a partir de um anterior (exemplo  o transformation filter). 
- As __actions__ calculam resultados a partir de um RDD.

As transformaes (transformations) so sempre postergadas at encontrar uma ao (action). O conceito de avaliao tardia(Lazy evaluation) em conjunto com os DAGs fazem com as otimizaes de cdigo no Spark sejam muito boas.

Para reusar o RDD em vrias 'actions' o comando 'persist' deve ser explicitado H algumas opes para persistncia dos dados. Uma vez que o 'persist'  utilizado, no haver mais reclculos quando houver outras 'actions. Pragmaticamente,  necessrio persistir os dados em memria uma vez e fazer vrias consultas ('actions') nestes dados em memria;



__Exemplos de transformations__:

- Map() : aplica uma funo a cada elemento do RDD, e os elementos gerados pela funo formam o novo RDD
- Filter(): aplica uma funo e retorno o conjunto de elementos filtrados com um novo RDD
- FlatMap() : Similar ao Map com a diferena que para cada item de entrada pode ser gerada um mais itens de sada;
- Distinct() = para gerar elementos distintos de um RDD;
- Union() = faz unio dos dados dos conjuntos. As repeties so mantidas;
- Intersection() = retorna os elementos existente em ambos os RDDs (elimina os duplicados = atua como o distinct(), mantendo sua obsevao quanto ao desempenho);
- Subtract() = retorna os valores existentes apenas no primeiro RDD e que no existem no segundo RDD (tome cuidado com o desempenho);
- Cartesian() = realiza um produto cartesiano entre os dois RDDs envolvidos.

__Exemplos de actions__:

- reduce(): Aplica uma funo em dois elementos (tipo de dados) do RDD gerando um resultado do mesmo tipo;
- fold(): Faz a mesma coisa que o Reduce, mas atribui um valor inicial para cada elemento na soma;
- collect(): Forma mais simples de obter todo contedo de um RDD;
- take(n): Retorna os n elementos do RDD;
- takeSample (withReplacement, num, seed): Faz uma amostragem com ou sem reposio (assume uma distruibuio uniforme dos dados);
- foreach(): executa uma funo em cada elemento do RDD, mas no retorna nada para o Driver;
- count(): retorna a quantidade de elementos do RDD;

#### Spark SQL
 a proposta Spark para lidar com dados estruturados e semi-estruturados.  capaz de carregar vrios tipos de dados(Json, Hive, Parquet)

__As consultas so realizadas utilizando SQL (atravs de um programa spark ou atravs conexes externas: JDBC/ODBC).__

Ainda possui forte integrao para ser utilizada com as linguagens spark: 
- Java
- Scala
- Python

Internamente os _SchemaRDDs_ so um conjunto de objetos do tipo _Row_ e cada objeto do tipo _Row_  uma espcie de vetor com campos definidos pelo _Schema_

#### DataFrame
 um tipo de objeto com vrios mtodos para manipular estruturas de dados com linhas e colunas (tabelas). Pode ser usado por todas as linguagens que acessam o Spark



Algumas operaes comuns:

- Tipo de dados das colunas: _DF.printSchema()_
- Obteno das primeiras linhas do data frame: _DF.head(5)_
- Nmero de linhas: _DF.count()_
- Nmero de colunas: _len(DF.columns)_
- Estatsticas descritivas para colunas numricas: _DF.describe()_
- Mostrando somente algumas colunas: _DF.select('Col1','Col2').show(5)_
- Quantidade de elementos distintos em uma coluna: _DF.select('Col1').distinct().count()_
- Eliminado linhas duplicadas em um data frame: _DF.select('Col1','Col2').dropDuplicates().show()_
- Excluindo linhas com valores nulos: _DF.dropna()_
- Preenchendo valores nulos com outros valores: _DF.fillna(-1).show(2)_
- Fitlrar linhas: _DF.filter(DF.Col1 > 15000)_
- Gerao de tabelas cruzadas(crosstab): _DF.crosstab('Col1', 'Col2').show()_
