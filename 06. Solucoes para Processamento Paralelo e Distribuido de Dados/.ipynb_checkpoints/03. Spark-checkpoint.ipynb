# Spark

 uma plataforma (em cluster) para processamento rpido, paralelo e de propsito geral que extende o algoritmop adro: MapReduce. Muitos clculos so feitos em memria (podem ser executados no disco tambm).

Possui APIs para:
- Java
- Scala
- Python
- SQL e
- R.



Conceitos importantes em uma aplicao Spark:

- Driver Program: Qualquer aplicao Spark  um 'Driver program'. Esta aplicao  capaz de fazer ope raes paralelas no cluster Spark.  como se fosse uma funo principal (main) e define/criadados distribudos no cluster e aplica operaes paralelas aos dados distribudos no cluster. As operaes no so apenas Map e Reduce;
- Contexto Spark: O 'Driver program' acessa o spark atravs de um objeto contexto que na verdade, representa a conexo da aplicao com o Cluster. Possui alguns parmetros de configurao.
- Executors: Para realizar as operaes do 'Driver Program', o mesmo faz uso de vrios ns que possuem os executores
