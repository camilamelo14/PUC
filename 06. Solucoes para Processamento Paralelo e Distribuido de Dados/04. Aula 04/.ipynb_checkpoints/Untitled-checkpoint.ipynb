Duas funções são muito comuns são:

- Map() : Aplica uma função a cada elemento do RDD, e os elementos gerados na função forman um RDD
- Filter(): Aplica uma função e retorna o conj. de elementos filtrados como um novo RDD



A função Map é muito utilizada para tratar item a item aplicando uma função no conj. e o tipo de RDD de entrada não precisa ser o mesmo da saída.

- FlatMap(): É similar ao Map(), mas dif. retorna um iterador como resultado final.

### Transformação de conjunto

- Distinct(): Para gerar elementos distintos
    Operação custosa -> Pois utiliza-se de um processo de ordenação
- Union(): Faz o união dos dados dos conjuntos. As repetições são mantidas
- Intersection(): Retorna elementos em ambos os RDDs -> Elimina os duplicados e é custoso
- Subtract(): Retorna valores existentes apenas no primeiro RDD e que não existe no segundo. -> Tome nota do desempenho também





- parallelize() - Não mt usado pois precisa caber na memória de uma mesma máquina

### Ações - Spark

Básico:

-reduce(): Recebe uma lista de dados e retorna um resultado de um RDD com o mesmo tipo (Utilizado geralmente para agregar elementos).
-fold(): Similar ao Reduce porém atribui um valor inicial para cada elemento;
-aggregate(): Existe para auxiliar na função reduce
