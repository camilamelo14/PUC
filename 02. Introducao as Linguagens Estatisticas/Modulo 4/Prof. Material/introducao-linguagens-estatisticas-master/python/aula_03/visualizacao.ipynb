# Matplotlib

Matplotlib é uma biblioteca Python para traçar gráficos 2D que produz figuras de qualidade em vários formatos distintos. Matplotlib pode ser usado em scripts Python, no Python e IPython shell, no jupyter notebook entre outros.

Matplotlib é uma ferramenta bem versátil, que nos permite fazer praticamente qualquer tipo de gráfico (nem todos de maneira fácil). Pode-se traçar pontos, histogramas, gráfico em barras, gráficos de dispersão, etc., com poucas linhas de código ([galeria](https://matplotlib.org/gallery.html)).

Praticamente, todas as bibliotecas de visualização estatística, tais como seaborn, plotnine entre outras, são construídas utilizando rotinas do Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

%matplotlib inline

np.random.seed(123)
```

```python
X = np.abs(np.random.normal(60, 500, 100))[:, np.newaxis] 
X[X <= 40] = 40
w = 2000
y = X * (w - np.random.normal(0, 500, X.shape[0]))[:, np.newaxis] + 60000

plt.scatter(X, y)

# plt.show()
```

```python
X_l = np.hstack((np.ones((X.shape[0], 1)), X))

b, w = np.linalg.inv(X_l.T.dot(X_l)).dot(X_l.T).dot(y).ravel()

plt.scatter(X, y)
plt.plot(X, w*X + b, color="green")
```

```python
mu = 5.06
varianca = 0.124
sigma = np.sqrt(varianca)
x = np.linspace(mu - 3*varianca, mu + 3*varianca, 100)

# plotando uma distribuição normal
plt.plot(x, stats.norm(mu, sigma).pdf(x))
```

### Primeiro Modelo de Classificação Automática

```python
X, y = iris[iris.columns[:-1]].as_matrix(), iris[iris.columns[-1]].as_matrix()

import matplotlib as mpl

def make_ellipses(means, variances, ax, color):
    """Adaptação do código encontrado em:
    http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_covariances.html
    Para entender melhor como funciona olhar seguinte explicação:
    http://www.visiondummy.com/2014/04/draw-error-ellipse-representing-covariance-matrix/
    """
    for c in range(variances.shape[0]):
        # como as variáveis são independente
        # a matrix de covariância é uma matriz
        # diagonal, onde a diagonal são as 
        # variâncias dos atributos
        covariance = np.diag(variances[c,:])
        v, w = np.linalg.eigh(covariance)
        u = w[0] / np.linalg.norm(w[0])
        angle = np.arctan2(u[1], u[0])
        angle = 180 * angle / np.pi  # convert to degrees
        v = 2. * np.sqrt(5.991) * np.sqrt(v)
        ell = mpl.patches.Ellipse(means[c], v[0], v[1],
                                  180 + angle, color=color[c])
        ell.set_clip_box(ax.bbox)
        ell.set_alpha(0.5)
        ax.add_artist(ell)
        
current_palette = sns.color_palette("husl", 8)

colors = [current_palette[1], current_palette[3], current_palette[6], current_palette[5]]
```

Nós queremos saber qual é a probabilidade de uma determinada flor de iris (determinada por suas característicaas) pertencer a alguma das três espécies, matematicamente:

$\hat{E} = \arg \max P(E | M)$

onde 
- M é o vetor de medidas ($m_{sl}, m_{sw}, m_{pl}, m_{pw}$)
- E representa uma das possíveis espécies {setosa, versicolor, virginica}
- $\hat{E}$ é a espécie predita para as dadas medidas

![iris](dados/iris_petal_sepal.png)

Como nós vimos previamente com nossas visualizações que $P(m_i | E)$ (a probabilidade de um valor de uma das medidas acontecer dado uma espécie E) é mais ou menos uma normal $N(\mu_{m_i}, \sigma_{m_i}|E)$ $\forall m_i \in \{m_{sl}, m_{sw}, m_{pl}, m_{pw}\} \text{ e } E \in \{setosa, versicolor, virginica\}$. 

Aplicando o [teorema de bayes](https://pt.wikipedia.org/wiki/Teorema_de_Bayes) temos que:

$P(E | M) = \frac{P(E)P(M|E)}{P(M)}$

Como $P(M)$ é constante podemos ignorá-lo, portanto:

$P(E | M) \propto P(E)P(M|E)$

Lembre-se que $P(M|E) = P(m_{sl}, m_{sw}, m_{pl}, m_{pw}|E)$. Assumindo que as variáveis são independentes temos que:

$P(m_{ls}, m_{ws}, m_{lp}, m_{wp}|E) = P(m_{sl}|E)P(m_{sw}|E)P(m_{pl}|E)P(m_{pw}|E)$

Portanto, 

$P(E | M) = P(E)P(m_{sl}|E)P(m_{sw}|E)P(m_{pl}|E)P(m_{lw}|E)$

$ = P(E)N(\mu_{m_{sl}}, \sigma_{m_{sl}}|E)N(\mu_{m_{sw}},\sigma_{m_{sw}}|E)P(\mu_{m_{pl}}, \sigma_{m_{pl}}|E)P(\mu_{m_{lw}},\sigma_{m_{lw}}|E)$


Então, precisamos calcular $P(E) \forall E \in \{setosa, versicolor, virginica\}$ e as médias e variancias dada cada valor deEs.

```python
def calcular_medias_variancas(df):
    """ 
    Função calcula estatísitcas
    necessárias para predição
    -------
    params:
        df pandas.DataFrame
    """
    g = df.groupby('species')
    return (g.mean(), # média de cada variável por espécie
            g.std()**2, # variancia de cada variável por espécie
            df.species.value_counts()/df.shape[0]) # P(E), probabilidade de ser pertencer a cada espécie


medias, variancas, P_E = calcular_medias_variancas(iris)

# só estou renomeando as colunas
# para ficar parecido com a nomenclatura
# da explicação (não se preocupe com esse trecho)
medias.columns = ["m_%s" % "".join([v[0] for v in c.split()]) for c in medias.columns]
medias
```

```python
# só estou renomeando as colunas
# para ficar parecido com a nomenclatura
# da explicação (não se preocupe com esse trecho)
variancas.columns = ["v_%s" % "".join([v[0] for v in c.split()]) for c in variancas.columns]
variancas
```

```python
P_E
```

```python
import numpy as np
def predizer(X, medias, variancas, P_E):
    """ 
    Prediz espécies de cada instância de X
    dado as estatísticas previamente computadas
    -------
    params:
        X numpy.ndarray onde cada linha
        é uma flor iris e a coluna uma 
        das medidas.
        medias pandas.DataFrame onde cada
        linha contém as médias das medidas
        de cada espécie.
        variancias pandas.DataFrame onde cada
        linha contém as variancias das medidas
        de cada espécie.
        P_E pandas.Series probabilidade de cada
        espécie ocorrer.
    """
    # extraindo as espécies e quantidade
    especies, n_especies = P_E.index, P_E.shape[0]
    
    # criando uma matriz para as predições de cada classe
    pred = np.zeros((X.shape[0], n_especies))
    
    # calculando P(E)P(m_sl|E)P(m_sw|E)P(m_pl|E)P(m_lw|E)
    # para cada espécie E
    for i, especie in enumerate(especies):
        P_mi_dado_E = stats.norm(medias.loc[especie], np.sqrt(variancas.loc[especie])).pdf(X)
        pred[:, i] = P_E[especie] * np.prod(P_mi_dado_E, axis=1)
    
    id_y = np.argmax(pred, axis=1)
    return especies[id_y]
        
E_pred = predizer(X, medias, variancas, P_E)
E_pred
```

```python
# calculando a taxa de acerto
acerto = np.mean(y == E_pred) * 100
print("Taxa de acerto de {}%".format(acerto))
```

```python
from itertools import combinations

means = medias.as_matrix()
variances = variancas.as_matrix()

combs = combinations(range(X.shape[1]), 2)
fig, axs = plt.subplots(ncols=X.shape[1], nrows=X.shape[1], figsize=(20, 15))

for f1, f2 in combs:

    for i, c in enumerate(np.unique(y)):
        mask = y == c
        axs[f1, f2].scatter(X[mask, f1], X[mask, f2], color=colors[i])
        axs[f2, f1].scatter(X[mask, f2], X[mask, f1], color=colors[i])

    make_ellipses(means[:, [f1,f2]], variances[:,[f1,f2]], axs[f1, f2], colors)
    make_ellipses(means[:, [f2,f1]], variances[:,[f2,f1]], axs[f2, f1], colors)
```
