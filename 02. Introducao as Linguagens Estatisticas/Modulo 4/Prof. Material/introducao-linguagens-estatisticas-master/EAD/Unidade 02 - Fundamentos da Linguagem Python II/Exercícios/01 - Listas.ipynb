# Listas 

Considere uma lista vazia (list = []). Você pode executar os seguintes comandos:

1. **insert i e*: Insireir o inteiro $e$  na posição $i$.

2. **print**: Imprima a lista.

3. **remove e**: Delete a primeira ocorrência do inteiro $e$.

4. **append e**: Insira o inteiro $e$ no final da lista.

5. **sort**: Ordene a lista.

6. **pop**: Remove o último elemento da lista.

7. **reverse**: Reverta a lista.

## Tarefa

Inicialize sua lista e leia o valor de $n$ que será seguido por $n$ linhas de comandos onde cada comando será uma dos 7 tipos listados acima. Itere sobre cada comando em ordem e execute a operação correspondente na lista.

### Formato de Entrada

A primeira linha contém um inteiro, $n$ , que denota a quantidade de comandos. 
Cada linha $i$ das $n$ subsequentes linhas contêm um comando como descrito acima.

### Restrições

Os elementos adcionados na lista devem ser inteiros.

### Formato de Saída

Para cada comando do tipo **print**, imprima a lista em uma nova linha.

### Exemplo de Saída 0

```
12
insert 0 5
insert 1 10
insert 0 6
print
remove 6
append 9
append 1
sort
print
pop
reverse
print
```

### Exemplo de Saída 0

```
[6, 5, 10]
[1, 5, 9, 10]
[9, 5, 1]
```  

## Submissão do Código

Para submissão e codificação da tarefa utilizar o link a seguir: https://www.hackerrank.com/contests/ead-fundamentos-python-ii/challenges/python-lists. Essa paǵina é apenas a tradução das instruções do link anterior.
