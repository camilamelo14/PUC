# Detectando Números de Pontos Flutuantes com Regex

## Tarefa

A você é dado uma string $N$. Sua tareda é verificar se $N$ é um número de ponto flutuante.

Nessa tarefa, um número de ponto flutuante v´alido deve satisfazer todos os seguintes requisitos:

O número pode iniciar com os símbolos de **+**, **-** ou **.**. 

#### For example: 

✔ +4.50 

✔ -1.0 

✔ .5 

✔ -.7 

✔ +.4 

✖ -+4.5

O número deve conter pelo menos um valor decimal. 

#### For example: 
✖ 12. 
✔ 12.0  

O número deve ter exatamente um símbolo **.** . 
O número não deve dar nenhum erro quando convertido usando ```float(N)```.

### Formato de Entrada

A primeira linha contém um número inteiro $T$, indicando o número de casos de testes. 

As próximas $T$ linhas contêm uma string $N$ cada.

### Formato de Saída

Imprima, para cada caso de teste, **True** se o número for float ou **False** caso contrário.

### Exemplo de Entrada 0

``` 
4
4.0O0
-1.00
+4.54
SomeRandomStuff
```

### Exemplo de Saída 0

```
False
True
True
False
```

### Explicação 0

4.0**O**0 : **O** não é um digito. 

-1.00: é válido. 

+4,54: é válido. 

SomeRandomStuff: não é um número.

## Submissão do Código

Para submissão e codificação da tarefa utilizar o link a seguir: https://www.hackerrank.com/contests/ead-fundamentos-python-ii/challenges/introduction-to-regex. Essa paǵina é apenas a tradução das instruções do link anterior
