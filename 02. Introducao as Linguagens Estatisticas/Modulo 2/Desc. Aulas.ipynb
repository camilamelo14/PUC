<b> Trabalhando com as posições da lista

```python
s = ["1", "2", "3", "4", "5", "6"]

# acessando o primeiro e o último elemento
print(s[0])
print(s[-1])

# imprime a lista sem o 
# primeiro e último elementos
print(s[1:-1])
print(s[:-1])

# acessando elementos de n em n, no exemplo de 2 em 2 iniciando a primeira posição
print(s[::2])

# 1 em 1 iniciando a terceira posição
print(s[3::-1])

#Inversão de lista/string
nome = 'Camila'
print(nome[::-1])
```

```python
# Exemplos de manipulação de string com python
string = "Isso é apenas um simples exemplos de string"

# função len retorna a quantidade de caractéres,
# ou em outras palavras, o tamanho da string.
print("tamanho da string: ", len(string), '\n')

# Número de ocorrências de uma substring
# com a função `count`
print("número de i's : ", string.count("i"))
print("número de in's : ",string.count("in"), '\n')

# Primeira ocorrência de uma substring
# com a função `find`
print("O primeiro `i` está na posição : ", string.find("i"))
print("O primeiro `in` está na posição : ", string.find("in"), '\n')

# Separar uma string com split
print("Separação por espaço em branco: \n", string.split())
print("Separação por a's : \n", string.split("a"), '\n')

# Manipulando capitalização das strings.
# Isso é muito simples em python com as
# funções upper, lower and title.

# Para deixar tudo maiúsculo
print("UPPER:", string.upper())

# Para deixar tudo maiúsculo
print("lower:", string.lower())

# Para deixar tudo camel-case
print("Title:", string.title(), '\n')


# Lembre-se que strings são como listas de caracteres
# portanto é possível acessá-lo como acessamos
# os elementos de uma lista, veja os exemplos a seguir:

# acessando o primeiro caracter
print(string[0])

# acessando o último caracter
print(string[-1])

# acessando uma substring
print(string[1:-1])

# pegando dos os caracteres em posições pares
print(string[::2])

# invertendo uma string
print(string[::-1])
```

<b>Expressão Regular

Dito em aula que também existe o () sig. agrupamento

```python
import re

string = "5555555.01 + 2.122"
pattern = "[0-9]*\.[0123456789]*"
print(re.findall(pattern,string))

string = "5555555.01 + 2.122 + 5. + .0 + 10.2"
#Uso de + para forcar a ocorrência
pattern = "[0-9]+\.[0123456789]+"
print(re.findall(pattern,string))

string = "5555555.01 + 2.122 + 5. + .0 + 10.2 + 5"
#Uso de + para forcar a ocorrência
pattern = "([\d]+\.[\d]+)|([0-9])"
print(re.findall(pattern,string))

string = "5555555.01 + 2.122 + 5"
# Padrao do inicio ^ e do final $
pattern = "^([0-9]+\.[\d]+)"
print(re.findall(pattern,string))

text = "Camila Melo é legal e o Rafael é chato"
pattern = "([A-Z][a-z]+\s[A-Z][a-z]+)|([A-Z][a-z]+)"
for macth in re.finditer(pattern,text):
    print(macth)
    
text = "Camila Melo é legal e o Rafael é chato Marcos"
pattern = "([A-Z][a-z]+\s*)+"
for macth in re.finditer(pattern,text):
    print(macth)
```

<b>Expressão Regulares

Link: https://regex101.com/

```python
# Expressões regulares em Python

# importando o módulo de expressão regular
import re

string = "Teste de expressão regular em Python 3."

# procurando por um padrão
resultado = re.search('regular', string)

# capturando o resultado da pesquisa
print(resultado.group(0))

# caso nada seja encontrado o resultado seria None
resultado = re.search('nada', string)

if resultado:
    print("Encontrado")
else:
    print("Não encontrado")

# os padrões são organizados em grupos
resultado = re.search('(reg)(ular)', string)
#Com os parenteses pode ser agrupado
print("grupo 1:", resultado.group(1))
print("grupo 2:", resultado.group(2))

# Passando zero para o método retorna ambos os grupos.
print("grupo 0:", resultado.group(0))

# função match
resultado =  re.match('regular', string)

# .
resultado = re.match('...................(regular)', string)

# substring do inicio até a palavra `regular`
print(resultado.group(0))

# apenas a palavra regular
print(resultado.group(1), '\n')

# podemos obter os mesmos resultados
# utilizando ferramentas apropriadadas
resultado = re.match('.{19}(regular)', string)
print(resultado.group(0))

# apenas a palavra regular
print(resultado.group(1), '\n')

resultado = re.match('.*(regular)', string)
print(resultado.group(0))

# apenas a palavra regular
print(resultado.group(1), '\n')


print(re.findall("((reg)(ular))", string))
```

```python
import re
list = []

for i in range(int(input())):
    comando = input()
    
    operacao = comando.split()
    
    if re.search("insert",operacao[0]) != None: 
        list.insert(int(operacao[1]),int(operacao[2]))
    if re.search("print",operacao[0]) != None:
        print(list)
    if re.search("remove",operacao[0]) != None:
	    list.remove(int(operacao[1]))
    if re.search("append",operacao[0]) != None:
        list.append(int(operacao[1]))
    if re.search("sort",operacao[0]) != None:
        list.sort()
    if re.search("pop",operacao[0]) != None:
        list.pop(operacao[1])
    if re.search("reverse",operacao[0]) != None:
        list.reverse()
```
