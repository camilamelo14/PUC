###### Modelo Vetorial

Criado em 1968, proposto otiginalmente para resolver problemas de busca.

- Todo componente do sistema são vistos como conj. de palavras
- Espaço Vetorial (São apresentados assim)
- Dimensão do espaço é dada pelo número de palavras distintas. (As coordenadas são calculadas por exemplo pela frequencia das palavras)

O IDF busca expressar a "importância" de uma palavra na coleção.

    N: Número total de doc. de uma coleção
    nt: Número de documentos onde a palavra t ocorreu
    
$Idf(t) = log (\frac{N}{n_t})$
