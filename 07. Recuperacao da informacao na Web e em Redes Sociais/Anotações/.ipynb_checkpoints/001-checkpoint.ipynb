###### Modelo Vetorial

Criado em 1968, proposto otiginalmente para resolver problemas de busca.

- Todo componente do sistema so vistos como conj. de palavras
- Espao Vetorial (So apresentados assim)
- Dimenso do espao  dada pelo nmero de palavras distintas. (As coordenadas so calculadas por exemplo pela frequencia das palavras)

O IDF busca expressar a "importncia" de uma palavra na coleo.

    N: Nmero total de doc. de uma coleo
    nt: Nmero de documentos onde a palavra t ocorreu
    
$Idf(t) = log (\frac{N}{n_t})$
