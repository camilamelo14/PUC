## 12. Arquitetura e Qualidade de Dados (2018)

### UNIDADE 01

Arquitetura de dados e exemplos, conceitos de dados, entendendo o DMBoK, Gestão e Governança de dados.

Introdução ao conceito de arquitetura de dados e de suas sub arquiteturas. Conceito de dados, informação, conhecimento e sabedoria. Introdução ao framework DAMA-DMBoK1.

O que é arquitetura: Conj de partes construídas de forma independente, qeu devem ser integradas entre si buscando a harmonia entre si e funcionalidade do todo.

Exemplos:
- Arqui. predial
- Arqui. urbana

A arquitetura Corporativa é formada por:

    Arquitetura de Dados
    Arquitetura de Processos
    Arquitetura de Negócios
    Arquitetura de Sistemas
    Arquitetura de Tecnologia
    




Arquitetura de dados serve para: 
- Estabelecer conceitos comuns para entendimento dos dados
- Representar requsitos
- Apresentar o relacionamento/integração dos dados

E são gerenciadas por: Políticas, Padrões, Procedimentos, etc.

### Subtipos de arquitetura de dados

- Transacionais: Banco de dados, ERP, Legados
- Informacionais: DW/BI, Marts
- BigData/ IOT
- Metadados com dados sobre os dados

**A arquitetura de dados SEMPRE será centrada nos objetivos de negócios e visando uma proposta de valor**





- **Data Owners**: Área de negócio responsáveis pelos dados
- **Data Steward** (ingestor dos dados): São pessoas atribuídas/responsáveis pelos dados atribuídos a área
- **Data Custodian**: Equipe de TI Responsável por gerenciar os dados, backup, repositórios etc.
- **Data Quality**: Equipe que analisa os dados para avaliar o grau de qualidade dos mesmos
