## 12. Arquitetura e Qualidade de Dados (2018)

### UNIDADE 01

Arquitetura de dados e exemplos, conceitos de dados, entendendo o DMBoK, Gesto e Governana de dados.

Introduo ao conceito de arquitetura de dados e de suas sub arquiteturas. Conceito de dados, informao, conhecimento e sabedoria. Introduo ao framework DAMA-DMBoK1.

O que  arquitetura: Conj de partes construdas de forma independente, qeu devem ser integradas entre si buscando a harmonia entre si e funcionalidade do todo.

Exemplos:
- Arqui. predial
- Arqui. urbana

A arquitetura Corporativa  formada por:

    Arquitetura de Dados
    Arquitetura de Processos
    Arquitetura de Negcios
    Arquitetura de Sistemas
    Arquitetura de Tecnologia
    




Arquitetura de dados serve para: 
- Estabelecer conceitos comuns para entendimento dos dados
- Representar requsitos
- Apresentar o relacionamento/integrao dos dados

E so gerenciadas por: Polticas, Padres, Procedimentos, etc.

### Subtipos de arquitetura de dados

- Transacionais: Banco de dados, ERP, Legados
- Informacionais: DW/BI, Marts
- BigData/ IOT
- Metadados com dados sobre os dados

**A arquitetura de dados SEMPRE ser centrada nos objetivos de negcios e visando uma proposta de valor**





- **Data Owners**: rea de negcio responsveis pelos dados
- **Data Steward** (ingestor dos dados): So pessoas atribudas/responsveis pelos dados atribudos a rea
- **Data Custodian**: Equipe de TI Responsvel por gerenciar os dados, backup, repositrios etc.
- **Data Quality**: Equipe que analisa os dados para avaliar o grau de qualidade dos mesmos
