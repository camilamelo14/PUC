## Estilos de aprendizado

#### 1. Aprendizado supervisionado (supervised learning)
Possui rtulos conhecidos, Modelo que faz previses e se auto-corrige quando h previses ruins, at atingir um bom desempenho.

Abordagens:
- Induo de regras (rvore de deciso)
- Regresso (Prever valores futuros)
- Redes Neurais Multicamadas (Tcnica que pode ser utilizada em qq tarefa da rea de ML + Muito flexvel e poderosa)
- Aaprendizado baseado em instncias (Vizinhos mais prximos)
- Mtodos baseados em Kernels

#### 2. Aprendizado no supervisionado (unsupervised learning)
O dado no  rotulado/No possui resultado conhecido. Modelo deduz estruturas a partir da entrega. (O modelo tenta deduzir)

Abordagens:
- Agrupamento ou clustering
- Deteco de anomalias (Detecta valores muitos dif. dos valores normais *outliers*
- Redes neurais geradoras adversrias (Generative Adversal Networks)

#### 3. Aprendizado semisupervisionado (semisupervised learning)
Mistura de dados rotulados (poucos) e no rotulados(muitos). -> Tecnica utilizada onde o custo para rtulo  muito caro, ex: exames.

Existe uma previso desajvel, mas o modelo precisa organizar estruturas.

Abordagens:
- Mtodos geradores
- Separao de baixa densidade (Tcnica que detecta dados que esto agrupados e dados que esto fora da distribio normal)
- Mtodos baseados em grafos (Relaes entre dados)
- Mtodos heursticos

#### 4.Aprendizado por reforo (reinforcement learning)
Agente computacional interage com um ambiente dinmico para atingir um objetivo, no possui rotulos, mas iterage com o ambiente recebendo feedback.

* Funo de utilidade: Funo que indica se a regra criada foi efetiva ou no

Medida de desempenho ou utilidade define se o agente est caminhando na direo correta, no espao de solues.

Abordagens:
- Mtodo de Monte Carlo
- Mtodo de diferenas temporais
