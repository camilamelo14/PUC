## Estilos de aprendizado

#### 1. Aprendizado supervisionado (supervised learning)
Possui rótulos conhecidos, Modelo que faz previsões e se auto-corrige quando há previsões ruins, até atingir um bom desempenho.

Abordagens:
- Indução de regras (Árvore de decisão)
- Regressão (Prever valores futuros)
- Redes Neurais Multicamadas (Técnica que pode ser utilizada em qq tarefa da área de ML + Muito flexível e poderosa)
- Aaprendizado baseado em instâncias (Vizinhos mais próximos)
- Métodos baseados em Kernels

#### 2. Aprendizado não supervisionado (unsupervised learning)
O dado não é rotulado/Não possui resultado conhecido. Modelo deduz estruturas a partir da entrega. (O modelo tenta deduzir)

Abordagens:
- Agrupamento ou clustering
- Detecçção de anomalias (Detecta valores muitos dif. dos valores normais *outliers*
- Redes neurais geradoras adversárias (Generative Adversal Networks)

#### 3. Aprendizado semisupervisionado (semisupervised learning)
Mistura de dados rotulados (poucos) e não rotulados(muitos). -> Tecnica utilizada onde o custo para rótulo é muito caro, ex: exames.

Existe uma previsão desajável, mas o modelo precisa organizar estruturas.

Abordagens:
- Métodos geradores
- Separação de baixa densidade (Técnica que detecta dados que estão agrupados e dados que estão fora da distribição normal)
- Métodos baseados em grafos (Relações entre dados)
- Métodos heurísticos

#### 4.Aprendizado por reforço (reinforcement learning)
Agente computacional interage com um ambiente dinâmico para atingir um objetivo, não possui rotulos, mas iterage com o ambiente recebendo feedback.

* Função de utilidade: Função que indica se a regra criada foi efetiva ou não

Medida de desempenho ou utilidade define se o agente está caminhando na direção correta, no espaço de soluções.

Abordagens:
- Método de Monte Carlo
- Método de diferenças temporais

- Descrição de dados: Análise Exploratório (Caracteriza e Compara os dados)
- Associação: Identifica regras que podem descrever algum padrão nos dados
- Classificação e previsão
- Clusterização e Segmentação: Agrupa por semelhança
- Análise de tendências e desvios: Avalia caracteristicas ou tendências ao longo do tempo, a difereça com previsão é que a análise deste item se baseia no histórico dos valores passados, avalia características sazonais etc. (Sem o uso de dados externos)

## Definição do Problema

Definir o problema que precisa/deseja ser resolvido.

- Alinhamento de variáveis: Um determinado problema pode estar alinhado com o que???

## Redefinição do problema

Tarefa empírica, técnica de fatos e julgamento (Um problema não pode ser definido com precisão se for dado atenção diretamente e apenas ao problema) Esta técnica te ajuda a avaliar variáveis externas para compreender se estão relacionadas ao problema de origem.

- Fatos: Circustâncias que causam diretamento o problema
- Julgamentos: Observações a serem disputadas ou decididos. -> Inferrencias possíveis que podem estar relacionado ao problema de origem.

### Suporte e confiança

**Hiperparâmetros do modelo**: São parâmetros do modelo, o professor comentou sobre o agrupamento dos itens para minimizar divergencias do modelo.

Este calculo avalia a porcentagem de ocorrencia de certa associação com relação ao dataset, dependendo do tamanho da base informações menores podem ter grande relevância.



A Confiança vai avaliar o grau de certeza em que o A implica em B.



Suporte mínimo e confiança mínima são hiperparâmetros das regras de associação.

Algotitmos de associação: Apriori (Má performance) e FPGrowth cria o conceito de frequencia - arvore com padrões frequentes.

Desvantagem: Muitas vezes uma regra de associação com suporte alto podem representar regras óbvias.

Lift: indica a força de uma regra sobre a coocorrênciaaleatória de seus antecedentes e consequentes.



## Classificação e previsão

Problema de classificação (Classifica de algo - geralmente classificação binária)
Problema de regressão (Retorna um valor/agregação)

A distribuição das bases de treino e de teste devem ser semelhantes estatísticamente.

O Professor comentou sobre validação cruzada para resolução de problemas de desbalanceamento.

### Critériosde escolha de algoritmos:

1. Problema de Classificação ou Regressão
2. Tipos de dados/features
3. Distribuição das classes balanceadas
4. Interpretabilidade dos resultados (Necessita de um entendimento posterior etc.)

Parâmetros versus hiperparâmetros

    – Parâmetrosde um modelo: variáveis ajustadas no processo de aprendizado.
    – Hiperparâmetros: precisam ser ajustados, mas não são aprendidos.

#### Hiperparâmetros: Autotunning ou grid search.

Ajuste de variável de folga, é um problema que precisa ser avaliado nesta atividade.



Precisamos ter cuidado com o balanceamento entre Precisão e Recisão. Dependendo do problema a ser estudado.

Quando há desbalanceamento de classe uma técnica de avaliação é realizar a acurária por classe.
