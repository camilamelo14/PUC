#### Estimação Pontual e Intervalar

##### Intervalar para uma Média

    - Pontual: É a forma com a qual se melhor aproxima de um parâmetro
    - Média: O Melhor estimador é a já conhecida média aritimética.
    
Para um determinado parâmetro $\mu$, populacional, o melhor estimado será:

$$
x = \frac{\sum_{i=1}^n xi}{n}
$$

##### Intervalar para uma Intervalar

Também conhecido como intervalos de confiança, utiliza-se dois termos na amostragem:

    - Nível de confiança
    - Margem de erro
    
Em estimação intervalar para a média populacioanl $\mu$ temos dois tipos de intervalos:

Variância populacional {$\sigma^2$} conhecida

Variância populacional {$\sigma^2$} não conhecida

Ambas as formas supõem que dados (variável aleatória) seguem uma distribuição <b>Normal - Quantitativa Contínua

#### Variância conhecida

Suponha uma variável aleatória X normalmente distribuída com média $\mu$ e variância $\sigma^2$, ou seja, X~N$(\mu,\sigma^2)$
O intervalo de confiân para a média, considerando que a variância é conhecida, será dada por: 



$Z_{\frac{a}{2}}$: valor do score padrão para um nível de confiança 100(1-a)%

$\sigma$: Desvio padrão populacional

$n$: tamanho da amostra observada no estudo

```r
install.packages('readxl')
library(readxl)
Nota_Alunos <- read_excel("Pratica/Nota de Alunos - Parte 1.xlsx")
Nota_Alunos
```

```r
IC_Media_Conhecida <- function(x, sigma, confianca){
    n <- length(x)
    x_barra <- mean(x)
    IC_inf = x_barra - qnorm((1-confianca)/2,0,1,lower.tail = F)*(sigma/sqrt(n))
    IC_sup = x_barra + qnorm((1-confianca)/2,0,1,lower.tail = F)*(sigma/sqrt(n))
    saida <- cbind(IC_inf,IC_sup)
    saida
}
```

```r
IC_C_Prova1 <- IC_Media_Conhecida(Nota_Alunos$Prova_1,3,.95)
IC_C_Prova1
```
