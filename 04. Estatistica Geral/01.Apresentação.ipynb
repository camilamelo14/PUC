#### O que é estatística

É um conj. de métodos para o planejamento de estudos ou experimentos, obtenção de dados e consequente organização, resumo, apresetação, análise, interpretação e elaboração de <b>conclusões baseadas nos dados

#### População
É a coleção completa de todos os elementos a serem estudados.

#### Censo
É um conjunto de dados obtidos de todos os membros da população

Parâmetro só pode ser obtido pelo Censo

#### Amostra
É um subconjunto de membros selecionados de uma população

Obtém o p

#### Tipos de variáveis

- Qualitativo: Não numérico
- Quantitativo: Numérico

#### Áreas da Estatística

- Estatística Descritiva: Descrição, Esploração e Comparação dos dados:Medidas de Tendências, Central, Medidas de Variabilidade, Medisdas de Associação
- Estatística Inferencial: Distribuição de Probabilidade, Teste de Hipóteses e Intervalos de Confiança
- Modelagem Estatística: Análise de regressão, séries temporais e estatística multivariada.

### Representação Tabular e Gráfica (Aula 1 v2)

#### Tabelas de frequencia

A frequencia é a quantidade de vezes em que uma det. catagoria/faixa de valores aparece em um conjunto de dados.

- Nos dá uma noção de tendência dos dados (Muito importante na área de estatistica descritiva)



O T significa intervalo fechado (não inclusivo) neste caso do numero 5

$f_a$ - Frequancia absoluta

$f_r$ - Frequência relativa (%) = $x/X * 100$

$F_a$ - Frequancia absoluta acumulada: Soma os valores dos intervalos anteriores, avaliar o acumulo dos valores

$F_r$ - Frequência acumulada relativa: Percentual (%) em relação do total da amostra das frequências acumuladas absolutas.

#### Tipos de gráficos

- Gráfico de colunas
  - Colunas, gráfico vertical, com categorias ou faixa de valores no eixo $x$ e percentoais ou quantidades no eito $y$
  - Barras,  gráfico horizontal, com categorias ou faixa de valores no eixo $y$ e percentoais ou quantidades no eito $x$
  - Uma boa opção para dados quantitativos discretos
  
  
  

- Gráfico de Setores ou Pizza
  - Mais utilizados para trabalhar com percentual (%)
  - Uma boa opção para dados qualitativos (com poucas categorias)
  
  
- Histogramas
    - Uma boa opção para dados quantitativos contínuos 
    

- Gráfico de linhas ou de séries
    - Pode ser utilizao para representar a variação ao longo do tempo
    - Análise de séries temporais
    - Normalmente o tempo no eixo $x$
    - Auxilia para avaliar o comportamento nas funções
