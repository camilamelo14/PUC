## Probabilidade

A probabilidade é uma área da matemática que avalia a incerteza. Na estatística, tem um papel fundamental para modelar essas incertezas sobre nossos conjuntos de dados

#### Conceitos sobre probabilidade

- Experimento aleatório: É a situação ou acontecimento cujos resultados não podem ser previstos com certeza
- Espaço amostral: É o conjunto de todos os resultados possíveis de um experimento aleatório, no caso de um dado as possibilidades são {1,2,3,4,5,6} exemplo.
- Eventos: Sõ subconjuntos do espaço amostral, são sempre representados por letras MAIÚSCULAS {A,B,C,D ...}
    Ex: Podemos definir como um evento, se após o lançamento de um dado, obtermos as faces 3 ou 5, ou seja, evento A = {3,5} 


#### Definição clássica

A probabilidade é a relação de números de pontos de eventos vs numero de pontos do espaço amostral. Ou Seja, se A é o evento de interesse a probabilidade é representada: $P(A)$:

$$ 
P(A) = \frac{Num-casos-favoraveis-evento-A}{Num-casos-possiveis}
$$

A probabilidade de aparecer a face 3,5 é:

$$ 
P(A) = \frac{2}{6} = \frac{1}{3}
$$

#### Definição frequentista

A probabilidade de um evento como a proporção de vezes que ele se repete.

$$ 
P(A) = \frac{Num-de-vezes-que-o-evento-A-ocorreu}{Num-total-de-repeticao-do-experimento}
$$

Exemplo: Jogamos um dados de 6 faces 10 vezes e procuramos a ocorrencia do evento A, resultado 1,2,3,3,4,4,5,6,6,6, a probabilidade de aparecer 5 ou 3 é:

$$P(A) = \frac{3}{10}$$



### Tipos de eventos

Quando ha dois eventos, pode-se relacionar da seguinte maneira:

#### Interseção
É a ocorrência simultânea de dois eventos $A \cap B$

#### Disjunto
Também conhecido como mutuamente exclusivo/excludentes, ocorre quando dois eventos não possuem nenhum elemento em comum $A \cap B = 0$

#### União
Quando ocorre pelo menos um dos eventos em A ou B: $A \cup B$

Caso exista algum valor que corresponde nos dois eventos estes são unificados

#### Complementar
É quando a junção dos dois eventos foram o espaço amostral e sua interseção é vazia.

$ A \cup A^c = \omega $ A união dos valores formam o espaço amostral

$ A \cap A^c = 0 $ A interseção dos dois valores é vazia

### Tipos de eventos



A diferenção entre a união e a interseção é que na união os todos os valores são complementados, na interseção apenas o que está em comum entre os eventos.

### Regras básicas de probabilidade

- Se A e B são dois eventos do espaço amostral $\omega$ então valem as seguintes regras:
    - a) A probabilidade deve estar entre 0 e 1: 0<= P(A) <= 1 e  0<= P(B) <= 1
    - b) A probabilidade da amostra será sempre 1: $P(\omega) = 1$
    - c) Se A e B são eventos disjuntos então: $P( A\cup B) = P(A) + P(B)$
    
- Com essas regras, temos:
$P(A \cup B) = P(A) + P(B) - P( A \cap B)$  - Retirado a interseção para o valor não ficar dorado na conta

- A outra relação:
$P(A^c) = 1- P(A)$

### Probabilidade Condicional

- A probabilidade do evento A ocorrer dado que o evento B já ocorreu: 
$$P(A|B) = \frac{P(A\cap B)}{P(B)}$$

###### Probabilidade de A ocorrer, dado que B ocorreu $P(A|B)$, realiza a fração de A com interseção com evento B, dividido pela probabilidade de B 

    Obs: É conhecida como probabilidade condicional desde que P(B) seja diferente de zero
    
    Ex: Qual a probabildiade de chover amanhã, dado que ontem também choveu ou não choveu etc.
    
    
- Da definição de probabilidade condicional, deduzimos que a regra do produto é definida por:

$$ P(A \cap B) = P(A|B) P(B)$$    

    
#### Conhecida também como regra de multiplicação. 

### Exemplo/Exercícios

1) Suponha que são lançados dois dados equilibrados e que se desejam calcular as seguintes probabilidades:
    
    a) Soma das duas faces ser 8;
    b) Soma das duas faces ser 8, sabendo que ocorre face 3 no primeiro dado.
    
R:

    A = {Soma 8}
    B = {Ocorrer a face 3 no primeiro dado} >> Condição
    
    a) Caso existam 36 resultados possíveis (dois dados de 6 posições) no espaço amostral e 5 no resultados possíveis (valores que podem somar 8) para o evento A, logo P(A) = 5/36
    
#### Definição frequentista
    
$$ 
P(A) = \frac{Num-de-vezes-que-o-evento-A-ocorreu}{Num-total-de-repeticao-do-experimento}
$$

$$ P(A) = \frac{5}{36} $$


        b) Note que agora precisamos considerar o evento B, probabilidade condicional 
$P(A|B)$:
        
$$
P = (A|B) = \frac{P(A \cap B}{P(B)}
$$

Como só temos 6 possivilidades de ter a face 3 no primeiro dado, temos $P(B) = \frac{6}{36}$ e apenas uma possibilidade de termos a soma 8 com a face 3 no primeiro dado. (Face 3 D1 e face 5 D2), temos:

$$
P (A \cap B) = \frac{1}{36}
$$

$$
P (A | B) = \frac{\frac{1}{36}}{\frac{6}{36}} = \frac{1}{36} x \frac{36}{6} = \frac{1}{6}
$$

Obs: Mantén a primeira fração e inverte a segunda (Multiplicação de fração)


### Independência de eventos

Eventos independentes que a ocorrencia ou não de B não altera a probabiblidade de A

$$P(A|B) = P(A), P(B) > 0,$$

Ou:

$$ P(A \cap  B) = P(A) P(B) $$

    Exemplo: Uma empresa produz peças em duas máquinas I e II, que podem apresentar desajustes com probabilidade de 0,05 e 0,10; respectivamente. 
    
        No início do dia caso o teste seja realizado a a máquina esteja fora do ajuste, esta ficará sem funcionar e não irá afetar as outras máquinas em produção.
        
        A empresa não pode ficar sem produzir, para isso vamos utilizar a estatistica para calcular esta informação.
        
        
Solução: Seja o evento $O_i$ e $i = 1,2$

Pelas informações disponíveis temos $P(O_i) = 0,95 $ e $P(O_2) = 0,90$

#### Diagrama da árvore de probabilidade

Como o funcionamento na máquina1 não depende da máquina2 e vice-versa, temos:

$$P(O_1 \cap O_2) = P(O_2|O_1) P(O_1) = P(O_1) P(O_2) $$

### Modelo de Probabilidade Discreta

#### Modelo Bernoulli: 

    - Considera-se este modelo quando a variável X atribui 0 ou 1 à ocorrencia de fracasso ou sucesso respectivamente, sua função: 
$P(X = 1) = p$  $P(X=0)=1-p$

Ou

$P(X =x) = p^x(1-p)^{1-x}, x=0,1$

    - A repetição de ensaios de Bernoulli independentes dá origem à mais importante variáve aleatória discreta denominada modelo Binomial
    
#### Modelo Binomial:    

    - Considere a repetição de n ensaios de Bernoulli indepentendes e todos com a mesma probabilidade p.
    - A variável aleatória que conta o número de sucesso é denominada Binomial com parâmetros n e p. sendo:
    
$P(X = k) = (\frac{n}{k})p^k(1-p)^{n-k}$ sendo k=0,1,2, ...,n

K = Quantidade que ganha-se

Onde: 
$$(\frac{n}{k}) = \frac{n!}{k!(n-k)!}$$

Usaremos a notação X~B(n,p) para indicar que X segue um modelo Binomial com parâmetros n e p.

Calculo de probabilidade para X~Bin(10;0,2)

    -Ensaios = 10;
    -Probabilidade do sucesso = 0,2
    -X ganhos

```r
#P(X>8) = p(X>=9) = 1-P(X<=8)
pbinom(8,10,.2,lower.tail = F) #NÃO INCLUI O ULTIMO CALCULO 1-P(X<=8) E SIM O PRIMEIRO P(X>8) ( MAS NÃO FICOU CLARO A APLICAÇÃO)

#P(X<3) = P(X <=2)
pbinom(2,10,.2)

#P(2<=x<=5) = P(X<=5) - P(X<=1)
pbinom(5,10,.2)-pbinom(1,10,.2)
```

K = 0,1,2 ... (valor não fixo)

### Modelo de Probabilidade Contínua

Neste modelo o X não é calculado em quantidade inteira.

#### Modelo Exponencial

- A variável X pode assumir qualquer valor acima de 0, sua função:

$$
f(x) = \lambda{e}^{\lambda{x}}
$$

Para valores de $x$ >= 0 e 1/$\lambda$ é o valor esperado (média) de X.
    - A variável aleatória exponencial é adequada para estudos que envolvem tempo de vida ou funções de taxa de falha ao longo do tempo
    Ex: Tempo de vida de algum equipamento até apresentar alguma falha
    
Notação: $X~Exp(\frac{1}{\lambda})$

##### Exemplos:

    - Cálculo da Probabilidade para X~Exp(1/2000), Ou seja, a variável terá um tempo de vida médio de 2000 horas.

```r
#P(X > 1500)
pexp(1500,1/2000,lower.tail = F) #Calculando de 1500h para mais ref. ao tempo de probabilidade de não dar problema

pexp(2000,1/2000) #Até 2000h

pexp(3000,1/2000) - pexp(2000,1/2000)
```

#### Modelo Normal

A variável aleatória X tem distribuição Normal com parâmetros $\mu$ e $o^2$ se sua função densidade de probabilidade é dada por:



1. $\mu$ é o valor esperado (média) de X
2. $\sigma^2$ é a variância de X (maior que zero);
3. $\pi$ constante igual a 3,14159
4. X = variável aleatória em que:  $-\infty > x < \infty$

Notação: X~N$(\mu,\sigma^2)$

##### Calculo da probabilidade para X~N(10;2)

```r
pnorm(8,10,2, lower.tail = F) # P(X > 8)

pnorm(3,10,2) # P(X < 3) = P(X < 2)

pnorm(5,10,2) - pnorm(2,10,2) # P(2<=X<=5) = P(X<=5) - P(X<= 1)
```
