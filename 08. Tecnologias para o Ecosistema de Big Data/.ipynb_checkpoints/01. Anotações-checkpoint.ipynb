### 08. Tecnologias para o Ecosistema de Big Data 

### Hierarquia de pasta

- Cada arquivo  considerado uma parte da tabela
- As pastas representam contextos e tabelas



OBS: O formato e a estrutura destes arquivos devem ser iguais* (Formato do arquivo e estrutura de colunas)

Os dados podem ser divididos em subpastas.

- Quanto maior aquivos existir, maior o overhead de leitura, ou seja, maior tempo gasto para leitura destes arquivos*

###### Vantagens 

- Velocidade para agregao e leitura
- Eficincia no armazenamento dos dados -> Compresso e Custo de processamento desacoplado. (quando os dados esto em um local, mas o servidor de processamento est em outro local)

###### Desvantagens

- Realiza apenas inseres (No atualiza ou remove registros) -> Necessrio reescrever os arquivos que esto no local armazenado quando existe um novo arquivo.
- No  recomendado para sistemas que exijam transformaes complexas de dados, ou quando h diversos JOINS nas tabelas.

###### Dicas 

- Atentar-se a capacidade da memria RAM para que os resultados intermedrios no processamento paralelo sejam suportados
- Subdiratrios facilitam o uso de computao distribuda. (Armazenamento em servidores diferentes)

### Arquivo orientado por linha

- Cada linha representa um registro
- Possui uma seprao clara entre as colunas

##### CSV ou TSV (Tab)
- Caracter especial separam as colunas 
- Textos so limitados por " " (Opcional)

##### JSON
-  um formato leve para troca de gastos
- Facil para mquina processar e gerar


##### Vantagens
- Processamento simples: Leitura sequencial, Facil acesso a linhas especficas etc.
- Fcil insero

##### Desvantagens
- Agregaes so custosas
- Dificil compresso
- Operaes colunares exigem a leitura de dados desnecessrios

### Arquivos Orientados por Colunas

Tambm conhecidos como Record Columnar File (RCFile)

- Quebra de linha definem a separao de colunas



Ao realizar uma agregao o resultado  muito mais rpido e com menos processamento, pois, ao encontrar a coluna que deseja realizar a agregao a operao no necessita acessar todos os registros em busca do item a ser calculado, basta apenas agregar a coluna desejada.

##### Vantagens
1. Compresso mais eficiente
2. Rpido para agregaes e operaes colunares

##### Desvantagens
1. Pouco eficiente para insero e emoo de registros
2. Nao eficiente quando h necessidade de agrupar vrias colunas

#### Formatos de compresso

- GZIP (GNU): baseado em algoritimo LZ77 e Huffman (Utiliza mais CPU, recomendado para dados que so acessados com menor frequencia)
- Snappy (Google): No tem objetivo de maxima compresso mais sim rapidez no processo. 
- LZ4 (OpenSource): similar ao Snappy

#### Recomendaes:

Avalie sempre os seguintes pontos:
- Capacidade de compresso;
- Escrita e leitura;
- Trfego na rede.

### Apache Avro, Orc e Parquet

Formatos binrios legveis por mquina (possuem algum tipo de compresso)

- #### Apache Avro (linha): 

Sistema de serializao de dados orientado por linha, oferece estrutura de dados rica, compacta e rpida.

    - Tipagem dinmica (O dado  acompanhado pelo esquema, ou seja, no necessrio uma codificao para descompactar o dado)
    - Evita a criao de tags
    - Mudanas so facilmente tratadas
    
- #### Apache Parquet e Orc (coluna):

O principal objetivo  aumentar a performance de i/o com suporte eficiente a compresso e codificao do esquema

#### Apache Parquet (Modelo Dremel)

Cada arquivo contm os valores para o conj. de linhas -> Por mais que seja orientado por coluna cada arquivo representa um conj. de linhas.

Eficiente em termos de escrita/leitura de disco para consultas em colunas especficas.



#### Orc

 a evoluo do modelo RCFile (Record Columnar File - RCFile) -> Arquivo orientado por coluna. Ele armazena vrios conj. de linhas em formato colunar, entretanto, tem alguns detalhes que so diferentes do RCFile, sendo:

- Possui uma pequena indexao que permite pular blocos de linhas
- Permite processamento paralelo
- Possui estatstica bsica sobre as colunas
