### 08. Tecnologias para o Ecosistema de Big Data 

### Hierarquia de pasta

- Cada arquivo é considerado uma parte da tabela
- As pastas representam contextos e tabelas



OBS: O formato e a estrutura destes arquivos devem ser iguais* (Formato do arquivo e estrutura de colunas)

Os dados podem ser divididos em subpastas.

- Quanto maior aquivos existir, maior o overhead de leitura, ou seja, maior tempo gasto para leitura destes arquivos*

###### Vantagens 

- Velocidade para agregação e leitura
- Eficiência no armazenamento dos dados -> Compressão e Custo de processamento desacoplado. (quando os dados estão em um local, mas o servidor de processamento está em outro local)

###### Desvantagens

- Realiza apenas inserções (Não atualiza ou remove registros) -> Necessário reescrever os arquivos que estão no local armazenado quando existe um novo arquivo.
- Não é recomendado para sistemas que exijam transformações complexas de dados, ou quando há diversos JOINS nas tabelas.

###### Dicas 

- Atentar-se a capacidade da memória RAM para que os resultados intermedários no processamento paralelo sejam suportados
- Subdiratórios facilitam o uso de computação distribuída. (Armazenamento em servidores diferentes)

### Arquivo orientado por linha

- Cada linha representa um registro
- Possui uma sepração clara entre as colunas

##### CSV ou TSV (Tab)
- Caracter especial separam as colunas 
- Textos são limitados por " " (Opcional)

##### JSON
- É um formato leve para troca de gastos
- Facil para máquina processar e gerar


##### Vantagens
- Processamento simples: Leitura sequencial, Facil acesso a linhas específicas etc.
- Fácil inserção

##### Desvantagens
- Agregações são custosas
- Dificil compressão
- Operações colunares exigem a leitura de dados desnecessários

### Arquivos Orientados por Colunas

Também conhecidos como Record Columnar File (RCFile)

- Quebra de linha definem a separação de colunas



Ao realizar uma agregação o resultado é muito mais rápido e com menos processamento, pois, ao encontrar a coluna que deseja realizar a agregação a operação não necessita acessar todos os registros em busca do item a ser calculado, basta apenas agregar a coluna desejada.

##### Vantagens
1. Compressão mais eficiente
2. Rápido para agregações e operações colunares

##### Desvantagens
1. Pouco eficiente para inserção e emoção de registros
2. Nao eficiente quando há necessidade de agrupar várias colunas

#### Formatos de compressão

- GZIP (GNU): baseado em algoritimo LZ77 e Huffman (Utiliza mais CPU, recomendado para dados que são acessados com menor frequencia)
- Snappy (Google): Não tem objetivo de maxima compressão mais sim rapidez no processo. 
- LZ4 (OpenSource): similar ao Snappy

#### Recomendações:

Avalie sempre os seguintes pontos:
- Capacidade de compressão;
- Escrita e leitura;
- Tráfego na rede.

Dados já comprimidos não é recomendado recomprimí-los.

### Apache Avro, Orc e Parquet

Formatos binários legíveis por máquina (possuem algum tipo de compressão)

- #### Apache Avro (linha): 

Sistema de serialização de dados orientado por linha, oferece estrutura de dados rica, compacta e rápida.

    - Tipagem dinâmica (O dado é acompanhado pelo esquema, ou seja, não necessário uma codificação para descompactar o dado)
    - Evita a criação de tags
    - Mudanças são facilmente tratadas
    
- #### Apache Parquet e Orc (coluna):

O principal objetivo é aumentar a performance de i/o com suporte eficiente a compressão e codificação do esquema

#### Apache Parquet (Modelo Dremel)

Cada arquivo contém os valores para o conj. de linhas -> Por mais que seja orientado por coluna cada arquivo representa um conj. de linhas.

Eficiente em termos de escrita/leitura de disco para consultas em colunas específicas.



#### Orc

É a evolução do modelo RCFile (Record Columnar File - RCFile) -> Arquivo orientado por coluna. Ele armazena vários conj. de linhas em formato colunar, entretanto, tem alguns detalhes que são diferentes do RCFile, sendo:

- Possui uma pequena indexação que permite pular blocos de linhas
- Permite processamento paralelo
- Possui estatística básica sobre as colunas
